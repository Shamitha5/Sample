## 1: The Central Limit Theorem

Try out generating and plotting the distribution of the sum of N random numbers. Can you show them approaching a Gaussian? Is there anything you can do to make it not do so?

```python
import scipy.stats as sps
import matplotlib.pyplot as plt
import numpy as np

r = sps.norm()
t = sps.cauchy()
N = 1000
rand = np.zeros(N)
sum_n = np.zeros(N)
sum_c = np.zeros(N)

for i in range(N):
    rand_n=r.rvs(N)
    rand_c=t.rvs(N)
    sum_n[i]=sum(rand_n)
    sum_c[i]=sum(rand_c)

plt.hist(sum_n, bins=100)
plt.title("Samples from Normal distribution")
plt.show()
plt.hist(sum_c, bins=100)
plt.title("Samples from Cauchy distribution")
plt.show()
```

## 2: Goodness-of-fit

Now consider the problem of checking if this sum _is_ Gaussian. Generate M random sums of N random numbers, and try out some goodness-of-fit tests such as: 
- the Kolmogorov-Smirnov
- Anderson-Darling
- Chi-square
- (Bonus: if you choose the distribution of each of the N summands, can you compute an analytical distribution of the sum so that you can use the likelihood ratio?)

```python
#Kolmogorov-Smirnov Test 
#if p value less than 0.05, not normal (rejects null hypothesis)

mu_n, sigma_n = np.mean(sums_n), np.std(sums_n)
mu_c, sigma_c = np.mean(sums_c), np.std(sums_c)

ks_n, p_val_n = sps.kstest(sums_n, "norm", args=(mu_n, sigma_n))
ks_c, p_val_c = sps.kstest(sums_c, "norm", args=(mu_c, sigma_c))

print("KS test (p value) for Normal dist: p value = ", p_val_n)
print("KS test (p value) for Cauchy dist: p value = ", p_val_c)
```

```python
#Anderson-Darling Test

anderson_n = sps.anderson(sums_n, dist = "norm")
anderson_c = sps.anderson(sums_c, dist = "norm")

print("Anderson test statistic for Normal dist: ", anderson_n.statistic)
print("Anderson test statistic for Cauchy dist: ", anderson_c.statistic)
```

```python
#Bonus: if you choose the distribution of each of the N summands, 
#can you compute an analytical distribution of the sum so that you can use the likelihood ratio?

mu=0
sig=1

sums = np.sum(np.random.normal(mu,sig,(M,N)), axis=1)

#theoretical normal distribution
x_vals = np.linspace(min(sums), max(sums), 100)
theoretical_pdf = sps.norm.pdf(x_vals, loc=N*mu, scale=np.sqrt(N)*sig)

plt.hist(sums, bins=50, density=True, label="Empirical")
plt.plot(x_vals, theoretical_pdf, label="Theoretical")
plt.xlabel("Sum of N normal random numbers")
plt.legend()
plt.show()

logl_sums = np.sum(np.log(sps.norm.pdf(sums, loc=N*mu, scale=np.sqrt(N)*sig)))

mu_th, sig_th = np.mean(sums), np.std(sums)
logl_th = np.sum(np.log(sps.norm.pdf(sums, loc=mu_th, scale=sig_th)))

logl_ratio = np.exp(logl_sums - logl_th)

print(f"Log likelihood under theoretical normal: {logl_sums}")
print(f"Log likelihood under fitted normal: {logl_th}")
print(f"Likelihood Ratio: {logl_ratio}")
```

### Power investigation!
Try plotting the _power_ to reject the hypothesis that M sums of N random numbers each is a Gaussian distribution at some significance level alpha. 

```python
#for KS test

def power_calc(M,N,alpha,trials):
    count = 0  #rejection count
    for i in range(trials):
        sums = np.sum(np.random.normal(0,1,(M,N)), axis=1)
        sums/= sums.std()
        
        ks_stat, p_val = sps.kstest(sums, "norm")
        
        if p_val < alpha:
            count += 1
    
    power = count/trials
    return power

M = np.arange(2,20)
N = 100
alpha = 0.05
trials = 1000

powers = [power_calc(i,N,alpha,trials) for i in M]

plt.plot(M,powers)
plt.show()
print(powers)
```

```python
#for the Anderson Darling test 
#critical values = [15, 10, 5, 2.5, 1]

def power_calc(M,N,alpha,trials):
    count = 0  #rejection count
    for i in range(trials):
        sums = np.sum(np.random.normal(0,1,(M,N)), axis=1)
        sums /= sums.std()
        result = sps.anderson(sums, "norm")
        
        if result.statistic > result.critical_values[2]:
            count += 1
    
    power = count/trials
    return power

M = np.arange(10,1000,10)
N = 100
alpha = 0.05   #result.critical_values[2]
trials = 100


powers = [power_calc(i,N,alpha,trials) for i in M]

plt.plot(M,powers)
plt.show()
print(powers)
```

```python
n = sps.norm()
c = sps.cauchy()

N = 1000  #no of random numbers
M = 100  #no of sums
bins=100

random_n = n.rvs((N, M))

sums_n = np.sum(random_n, axis=0)

print(random_n.shape)
print(sums_n.size)

plt.hist(sums_n, bins=bins)
plt.title("Samples from Normal distribution")
plt.show()
```

```python
#Mismodelling, doubling every other bin

bin_counts, bin_edges = np.histogram(sums_n, bins=bins)  
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

for i in range(0, len(bin_counts), 2): 
    bin_counts[i] *= 2
    
plt.bar(bin_centers, bin_counts)
plt.title("Histogram with Mismodelling in Half the Bins")
plt.show()
```

```python
#Chi-square Test
#when we do not know the affected bins  #when p value is less than 0.05, mismodelling is detected

expected_freq = np.diff(sps.norm.cdf(bin_edges, mu_n, sigma_n)) * len(sums_n)

#Normalizing expected frequencies so the sums match
scaling_factor = np.sum(bin_counts) / np.sum(expected_freq)
expected_freq *= scaling_factor

chi2_stat, p_value = sps.chisquare(bin_counts, expected_freq)

print("chi2 test (p value) for Normal dist:", p_value)
```

```python
#when we know the affected bins  #when p value is less than 0.05, mismodelling is detected

affected_bins = range(0, len(bin_counts), 2)  

affected_mask = np.zeros_like(bin_counts, dtype=bool)
affected_mask[list(affected_bins)] = True  # Marking affected bins as True

affected_counts = bin_counts[affected_mask]
affected_expected_freq = expected_freq[affected_mask]

#Normalizing expected frequencies so the sums match
scaling_factor = np.sum(affected_counts) / np.sum(affected_expected_freq)
affected_expected_freq *= scaling_factor


chi2_stat_known, p_value_known = sps.chisquare(affected_counts, affected_expected_freq)


print(f"Chi-Square Test (p-value) for known affected bins: {p_value_known}")
```

```python
#Power calculation

n = sps.norm()
N = 1000  #Number of random numbers
M = 100   #Number of sums
bins = 100
num_simulations = 100  #Number of simulations for 
alpha = 0.05  #Significance level

power_known = 0
power_unknown = 0

#Simulating multiple times
for _ in range(num_simulations):
    
    random_n = n.rvs((N, M))
    sums_n = np.sum(random_n, axis=0)
    sums_n /= sums_n.std()
    
    #Mismodelling: doubling every other bin
    bin_counts, bin_edges = np.histogram(sums_n, bins=bins)  
    expected_freq = np.diff(sps.norm.cdf(bin_edges)) * len(sums_n)
    
    #Doubling every other bin
    for i in range(0, len(bin_counts), 2):
        bin_counts[i] *= 2
        
    #Normalize expected frequencies so the sums match
    scaling_factor1 = np.sum(bin_counts) / np.sum(expected_freq)
    expected_freq *= scaling_factor1
    
        
    #When we do not know the affected bins (use all bins)
    chi2_stat, p_value_unknown = sps.chisquare(bin_counts, expected_freq)
    if p_value_unknown < alpha:
        power_unknown += 1

      
    #When we know the affected bins (perform test only on affected bins)    
    affected_counts = bin_counts[::2]
    affected_expected_freq = expected_freq[::2]
    
    #Normalize expected frequencies for affected bins
    scaling_factor2 = np.sum(affected_counts) / np.sum(affected_expected_freq)
    affected_expected_freq *= scaling_factor2
    
    chi2_stat_known, p_value_known = sps.chisquare(affected_counts, affected_expected_freq)
    if p_value_known < alpha:
        power_known += 1

#Power calculation
power_known /= num_simulations
power_unknown /= num_simulations


print(f"Power when knowing the affected bins: ", power_known)
print(f"Power when not knowing the affected bins: ", power_unknown)
```
